```python
import pandas as pd 
```

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("s.csv")
```

```python
df
```

```python
df = df.sample(50)
```

```python
df.shape
```

```python
colors = ['#1f77b4', '#d57a17', '#2ca02c', '#801a14', '#9467bd']
plt.ylabel("level")
plt.xlabel("age")
df['age'].value_counts().plot(kind='bar', color=colors)
plt.show()
```

```python
gender_count = df['gender'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(gender_count,
        labels=gender_count.index,
        autopct='%1.1f%%')

plt.title("gender Distribution")
plt.show()
```

```python
dept_count = df['branch'].value_counts()

plt.figure(figsize=(8,5))
dept_count.plot(
    kind='bar', color=['green','yellow','orange','blue','purple','brown'])


plt.title("Branch-wise Employee Count")
plt.xlabel("Branch")
plt.ylabel("Number of Student")
plt.show()

print("Branch with Highest Students:")
print(dept_count.idxmax())
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['age'], bins=10)
plt.title("age Distribution of students")
plt.xlabel("age")
plt.ylabel("degree")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['adaptability_score'],
            df['AI_fear_score'])

plt.title("adaptability vs AI")
plt.xlabel("adaptability (score)")
plt.ylabel("'AI_fear_score")
plt.show()
```

```python
degree_count = df['degree'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(degree_count,
        labels=degree_count.index,
        autopct='%1.1f%%')

plt.title("degree Distribution")
plt.show()
```

```python
dept_count = df['cgpa'].value_counts()

plt.figure(figsize=(8,5))
dept_count.plot(
    kind='bar', color=['green','yellow','orange','blue','purple','brown'])


plt.title("cgpa-wise student Count")
plt.xlabel("cgpa")
plt.ylabel("Number of Student")
plt.show()

print("cgpa with Highest Students:")
print(dept_count.idxmax())
```
